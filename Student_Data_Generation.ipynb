**ABM-based Student Data Generation**

```python
import pandas as pd
import random
from google.colab import files

# Number of student agents
NUM_STUDENTS = 400

students_data = []

# Loop from 1 to 400
for i in range(1, 401):
    # Fixed total marks
    class9_total = 1100
    class10_total = 1100
    class11_total = 1100
    class12_total = 1100

    # Obtained marks (ABM behavior)
    class9_obt = random.randint(500, 1000)
    class10_obt = random.randint(500, 1000)
    class11_obt = random.randint(450, 1000)
    class12_obt = random.randint(450, 1000)

    # CGPA calculation
    avg_percentage = (
        (class9_obt / class9_total) +
        (class10_obt / class10_total) +
        (class11_obt / class11_total) +
        (class12_obt / class12_total)
    ) / 4 * 100

    cgpa = round((avg_percentage / 100) * 4, 2)
    cgpa = min(cgpa, 4.0)

    students_data.append([
        i,
        class9_obt, class9_total,
        class10_obt, class10_total,
        class11_obt, class11_total,
        class12_obt, class12_total,
        cgpa
    ])

# Create DataFrame
columns = [
    "S.No",
    "class9_obt", "class9_total",
    "class10_obt", "class10_total",
    "class11_obt", "class11_total",
    "class12_obt", "class12_total",
    "cgpa"
]

df = pd.DataFrame(students_data, columns=columns)

# Save to Excel
file_name = "student_abm_data.csv"
df.to_csv(file_name, index=False)

# Download Excel file
files.download(file_name)

# Show first 5 rows
df.head()
```
